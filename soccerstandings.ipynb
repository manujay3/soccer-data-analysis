```python
import pandas as pd
import os
import datetime

from directory import data_dir
```

```python
df_leagues = pd.read_csv(data_dir + "/leagues.csv")
df_players = pd.read_csv(data_dir + "/players.csv", low_memory=False)
df_teams = pd.read_csv(data_dir + "/teams.csv")
df_fixtures = pd.read_csv(data_dir + "/fixtures.csv")
df_teamStats = pd.read_csv(data_dir + "/teamStats.csv")
df_teamRoster = pd.read_csv(data_dir + "/teamRoster.csv")
df_keyEventDescription = pd.read_csv(data_dir + "/keyEventDescription.csv")
df_status = pd.read_csv(data_dir + "/status.csv")
df_venues = pd.read_csv(data_dir + "/venues.csv")
df_standings = pd.read_csv(data_dir + "/standings.csv")
```

```python
mySeasonType_input = input("Please enter the season ID: ")
mySeasonType = int(mySeasonType_input)
df_standings = df_standings[df_standings['seasonType'].isin([mySeasonType])]
df_standings
```

```python
df_standings = pd.merge(df_standings, df_teams, how = "inner", left_on = ["teamId"], right_on = ["teamId"])
df_standings = pd.merge(df_standings, df_leagues, how = "inner", left_on = ["seasonType"], right_on = ["seasonType"])
df_standings
```

```python

df_standings = df_standings.drop(columns={'seasonType', 'year_x','year_y', 'leagueId_x', 'last_matchDateTime', 'location', 'name', 'shortDisplayName', 'color', 'alternateColor', 'logoURL', 'venueId','slug', 'year_x', 'seasonSlug', 'leagueId_y', 'midsizeName', 'leagueName', 'leagueShortName', 'slug', 'venueId', 'abbreviation', 'timeStamp', 'form'})

df_standings.dropna(axis=1, how='all', inplace=True)
col1 = df_standings.pop('seasonName')
col2 = df_standings.pop('displayName')

df_standings.insert(0, 'seasonName', col1)
df_standings.insert(2, 'displayName', col2)
df_standings.pop('teamId')

if not df_standings.empty:
    season_name = df_standings['seasonName'].iloc[0].replace(' ', '_').replace('/', '-')
    file_name = f"{season_name}_standings.csv"

    df_standings.to_csv(file_name, index=False)

    print(f"\nSuccessfully saved data to: {file_name}")
else:
    print("\nDataFrame was empty. No file was saved.")

# Display the final table

df_standings
```
