```python
#mySeasonType = 12654 #2024-2025 prem season
mySeasonType_input = input("Please enter the season ID (e.g., 12654): ")
mySeasonType = int(mySeasonType_input)
df_prem_teamStats = df_teamStats[df_teamStats['seasonType'].isin([mySeasonType])]

df_prem_teamStats
```

```python
df_prem_teamStats = pd.merge(df_prem_teamStats, df_teams, how = "inner", left_on = ["teamId"], right_on = ["teamId"])
df_prem_teamStats = pd.merge(df_prem_teamStats, df_leagues, how = "inner", left_on = ["seasonType"], right_on = ["seasonType"])

df_prem_teamStats
```

```python
df_prem_teamStats = df_prem_teamStats.drop(columns={'seasonType', 'teamId','updateTime', 'location', 'name', 'shortDisplayName', 'color', 'alternateColor', 'logoURL', 'venueId','slug', 'year', 'seasonSlug', 'leagueId', 'midsizeName', 'leagueName', 'leagueShortName', 'abbreviation'})

col1 = df_prem_teamStats.pop('seasonName')
col2 = df_prem_teamStats.pop('displayName')

df_prem_teamStats.insert(0, 'seasonName', col1)
df_prem_teamStats.insert(2, 'displayName', col2)
df_prem_teamStats.pop('teamOrder')

df_prem_teamStats.dropna(inplace=True)
df_prem_teamStats.drop_duplicates(ignore_index=True)

df_prem_teamStats
```
